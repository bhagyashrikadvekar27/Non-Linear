```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\admin\Downloads\china_gdp.csv")
df
```

```python
X = df['Year'].values
X = X.reshape(-1,1)
y = df['Value'].values
```

```python
print(type(X), type(y))
print(X.shape, y.shape)
```

```python
plt.scatter(X,y)
plt.xlabel("Year")
plt.ylabel("Value")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
poly2 = PolynomialFeatures(degree=2)
X2 = poly2.fit_transform(X)
X2

model2 = LinearRegression()
model2.fit(X2,y)
model2.score(X2,y)

model2.coef_

plt.scatter(X,y,label='Original',c='m')
plt.plot(X,model2.predict(X2),label='Predicted',c='g', marker='D')
plt.xlabel("Year")
plt.ylabel("Value")
plt.legend(bbox_to_anchor=(1.1,1))
plt.show()
```

```python
poly3 = PolynomialFeatures(degree=3)
X3 = poly3.fit_transform(X)

model3 = LinearRegression()
model3.fit(X3,y)
model3.score(X3,y)
```

```python
plt.scatter(X,y,label='Original',c='m')
plt.plot(X,model3.predict(X3),label='Predicted',c='g', marker='D')
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend(bbox_to_anchor=(1.1,1))
plt.show()
```
